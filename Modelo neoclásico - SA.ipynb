# El Modelo Neoclásico de Crecimiento con impuestos

## El modelo
Considere las preferencias:

\begin{equation*}
U_{0} = \sum_{t=0}^{\infty} \beta^{t} L_{t} \dfrac{c_{t}^{1-\theta} - 1}{1-\theta}
\tag{1}
\end{equation*}

donde $c_{t} = \frac{C_{t}}{L_{t}}$. 

Suponemos que la población crece a tasa $n$: 
$$L_{t+1} = (1 + n)L_t = g_L L_t $$ 

La función de producción estará dada por una función Cobb-Douglas con proceso técnico neutral de Harrod

\begin{equation*}
Y_t = K_{t}^{\alpha}(A_t L_t)^{1−\alpha}
\tag{2}
\end{equation*}

donde $$A_{t+1} = (1 + g)A_t = g_A A_t$$

Se considera también la existencia de un gobierno, que cobra impuestos sobre el consumo y la producción (con tasas de impuesto $\tau_{t}^{Y}$ y $\tau_{t}^{C}$), el que enfrenta la siguiente restricción presupuestaria

\begin{equation*}
G_{t} = \tau_{t}^{Y} Y_{t} + \tau_{t}^{C} C_{t}
\tag{3}
\end{equation*}

La restricción de recursos de la economía está dada por:

\begin{equation*}
Y_{t} = C_{t} + I{t} + G_{t}
\tag{4}
\end{equation*}

La ley de acumulación del capital es:

\begin{equation*}
K_{t+1} = I_{t} + (1-\delta)K_{t}
\tag{5}
\end{equation*}

Donde $\delta \in (0, 1)$ es la tasa de depreciación del capital. Suponga una dotación inicial de capital $K_0 > 0$

## Resolución del planificador central
El objetivo será encontrar la senda de capital y consumo óptimos a partir de un estado inicial con capital igual a $K_0$. Para esto utilizaremos el método "Shooting"

### Planteamiento
Normalizamos $A_0$ y $L_0$ al valor $1$, se tiene que
\begin{align*}
L_t &= g_{L}^{t} \\
A_t &= g_{A}^{t}
\end{align*}

- Las variables en forma intensiva son del tipo:
$$ \widehat{x}_{t} = \dfrac{X_t}{A_t L_t} = \dfrac{X_t}{g_{A}^{t} g_{L}^{t}} $$

- Ya que lo usaremos más adelante, el producto, por ejemplo, queda:
$$ \dfrac{Y_t}{A_t L_t} = \dfrac{K_{t}^{\alpha}(A_t L_t)^{1−\alpha}}{A_t L_t}  = \widehat{k}_{t}^{\alpha} $$

- Usando la definición del producto en $(2)$, la definición del gasto de la ecuación $(3)$, y despejando la inversión en la ecuación $(5)$, podemos reemplazar en la restricción de recursos de la economía $(4)$

\begin{align*}
\underbrace{K_{t}^{\alpha}(A_t L_t)^{1−\alpha}}_{Y_t} &= C_t  + \underbrace{[K_{t+1} - (1-\delta)K_t]}_{I_t} + \underbrace{[\tau_{t}^{Y} \underbrace{K_{t}^{\alpha}(A_t L_t)^{1−\alpha}}_{Y_t} + \tau_{t}^{C}C_t]}_{G_t} \\
K_{t}^{\alpha}(A_t L_t)^{1−\alpha}(1 - \tau_{t}^{Y}) &= C_{t}(1+\tau_{t}^{C}) + K_{t+1} - (1-\delta)K_t \text{, multiplicamos por } \frac{1}{g_{A}^{t} g_{L}^{t}} \\
\widehat{k}_{t}^{\alpha}(1-\tau_{t}^{Y}) &= \widehat{c}_t (1+\tau_{t}^{C}) + g_A g_L \widehat{k}_{t+1} - (1-\delta) \widehat{k}_{t} \text{, usando la definición antes vista} 
\end{align*}

- También podemos escribir la función de utilidad en términos intensivos. Primero notamos que $c_t = \widehat{c}_t g_{A}^{t}$, entonces:

\begin{align*}
u(c_t) &= \dfrac{(\widehat{c}_t g_{A}^{t})^{1-\theta} - 1}{1-\theta} \\
\Rightarrow U_0 &= \sum_{t=0}^{\infty} \beta^{t} g_{L}^{t} \dfrac{(\widehat{c}_t g_{A}^{t})^{1-\theta} - 1}{1-\theta}
\end{align*}

- Con estas dos expresiones, planteamos el problema de optimización con las variables en forma intensiva:

\begin{align*}
\max_{\{\widehat{c}_t, \: \widehat{k}_{t+1}\}_{t=0}^{\infty} } U_{0}&: \sum_{t=0}^{\infty} \beta^{t} g_{L}^{t} \dfrac{(\widehat{c}_t g_{A}^{t})^{1-\theta} - 1}{1-\theta} \\
\text{s.a } \; \widehat{k}_{t}^{\alpha}(1-\tau_{t}^{Y}) &= \widehat{c}_t (1+\tau_{t}^{C}) + g_A g_L \widehat{k}_{t+1} - (1-\delta) \widehat{k}_{t} \\
& \text{con }\widehat{k}_0 \text{ dado}
\end{align*}


### Condiciones de Primer Orden
- El Lagrangiano es:

$$ \mathcal{L}_0 = \sum_{t=0}^{\infty} \beta^{t} \left[ g_{L}^{t} \dfrac{(\widehat{c}_t g_{A}^{t})^{1-\theta} - 1}{1-\theta} + \lambda_t \Big\{ \widehat{k}_{t}^{\alpha} (1-\tau_{t}^{Y}) - \widehat{c}_t (1+\tau_{t}^{C}) - g_A g_L \widehat{k}_{t+1} + (1-\delta) \widehat{k}_t \Big\} \right] $$

- Las CPO de la solución interior generalizando para todo $t$, son:

\begin{align*}
\dfrac{\partial \mathcal{L}_t}{\partial \widehat{c}_t} &= g_{L}^{t} (\widehat{c}_t g_{A}^{t})^{-\theta} g_{A}^{t} - \lambda_t (1+\tau_{t}^{C}) = 0 \\ \\
\dfrac{\partial \mathcal{L}_t}{\partial \widehat{k}_{t+1}} &= -\lambda_t g_A g_L + \beta \lambda_{t+1} \left[ \alpha \widehat{k}_{t+1}^{\alpha-1}(1-\tau_{t+1}^{Y}) + (1-\delta) \right] = 0 \\ \\
\dfrac{\partial \mathcal{L}_t}{\partial \lambda_t} &= \widehat{k}_{t}^{\alpha} (1-\tau_{t}^{Y}) - \widehat{c}_t (1+\tau_{t}^{C}) - g_A g_L \widehat{k}_{t+1} + (1-\delta) \widehat{k}_t = 0 \\ \\
& t = 0,1,...
\end{align*}

- La condición de transversalidad es:
$$ \lim_{t\to\infty} \beta^{t} (\widehat{c}_t g_A^{t})^{-\theta} \widehat{k}_{t+1} = 0 $$

- De las dos primeras CPO se desprende la ecuación de ***Euler***
\begin{align*}
\lambda_t &= \dfrac{g_{L}^{t} g_{A}^{t} (\widehat{c}_t g_{A}^{t})^{-\theta}}{1+\tau_{t}^{C}} \\
\Rightarrow \left( \dfrac{ \widehat{c}_{t+1} g_A }{\widehat{c}_t} \right)^{\theta} &= \beta \left[ \alpha \widehat{k}_{t+1}^{\alpha-1} (1-\tau_{t+1}^{Y}) + 1 - \delta \right] \dfrac{1+\tau_{t}^{C}}{1+\tau_{t+1}^{C}}
\end{align*}

### Estado estacionario
- Denotaremos las variables en estado estacionario simplemente quitándoles el índice "$t$"
- Evaluando el estado estacionario del capital en la restricción de recursos (restricción del Lagrangiano), nos queda:
\begin{align*} 
\widehat{c}_t = \left[ \widehat{k}^{\alpha} (1-\tau^{Y}) + (1-\delta)\widehat{k} - g_A g_L \widehat{k} \right] \dfrac{1}{1+\tau^{C}} 
\end{align*} 
Por lo tanto, el consumo también encuentra un estado estacionario cuando $\widehat{k}$ se encuentra en él: $\widehat{c}$

- Evaluando el estado estacionario del consumo en la ecuación de ***Euler***, nos queda:
$$ g_{A}^{\theta} = \beta \left[ \alpha \widehat{k}^{\alpha-1} (1-\tau^{Y}) + 1 - \delta \right] $$

Despejando $\widehat{k}$, nos queda:
    $$ \widehat{k} = \left[ \dfrac{\alpha \beta (1-\tau^{Y})}{g_{A}^{\theta} + \beta \delta - \beta} \right]^{\dfrac{1}{1-\alpha}} $$
- Con esta expresión, el consumo de estado estacionario es:
$$ \widehat{c} = \dfrac{1}{1+\tau^C} \bigg\{ \left[ \dfrac{\alpha \beta (1-\tau^{Y})}{g_{A}^{\theta} + \beta \delta - \beta} \right]^{\dfrac{\alpha}{1-\alpha}} (1-\tau^{Y}) + (1 - \delta - g_A g_L) \left[ \dfrac{\alpha \beta (1-\tau^{Y})}{g_{A}^{\theta} + \beta \delta - \beta} \right]^{\dfrac{1}{1-\alpha}} \bigg\}$$

- El producto en forma efectiva también encuentra un estado estacionario, ya que es función de $\widehat{k}$:
$$ \widehat{y} = \widehat{k}^{\alpha} = \left[ \dfrac{\alpha \beta (1-\tau^{Y})}{g_{A}^{\theta} + \beta \delta - \beta} \right]^{\dfrac{\alpha}{1-\alpha}}$$

- El gasto del gobierno en forma efectiva también encuentra un estado estacionario:
$$ \widehat{g} = \tau^{Y} \underbrace{\widehat{k}^{\alpha}}_{\widehat{y}} + \tau^{C} \widehat{c} $$

### Shooting Algorithm


Note que dado un nivel de consumo inicial $c_0$, en conjunto con la ecuación de Euler y la restricción presupuestaria, se puede obtener una secuencia de consumo y capital. Es posible entonces, reducir el problema a encontrar el nivel de consumo inicial que, dadas las ecuaciones, genera una secuencia que converge al estado estacionario a través del *"saddle path"*. 

El algoritmo utilizado encuentra este nivel de consumo inicial utilizando la siguiente receta:

1. Dados dos valores extremos para el consumo $(c_L, c_H)$, encuentra el valor medio entre estos y lo utiliza como nivel de consumo inicial $c_0$
2. Obtiene la secuencia (de tamaño T) de consumo y capital generada dados los valores iniciales. Si en el proceso, el nivel de capital es negativo, se concluye que el nivel consumo inicial es muy alto, de manera que para la siguiente iteración, se actualiza $c_H$ a $c_0$ anterior. En caso de que el capital exceda el valor máximo posible de estado estacionario (evaluar despejar $k$ en la restricción presupuestaria para $c=0$), se concluye que el consumo inicial es muy bajo, de manera que para la siguiente iteración, se actualiza $c_L$ a $c_0$ anterior. En todos estos casos, se vuelve a la etapa 1
$$ k_{max} = \left[ \dfrac{1-τY}{g_A g_L +δ-1} \right]^{1/(1-α)} $$ 
3. Al finalizar las secuencias, si los últimos valores obtenidos son cercanos a los estados estacionarios (desvío porcentual respecto al estado estacionario menor a la tolerancia indicada), se ha encontrado la solución. En caso de que el último valor de la secuencia de capital sea menor al estado estacionario (el tamaño del desvío a la izquierda es mayor que la tolerancia dada), nuevamente se concluye que el nivel de consumo inicial es muy alto, en caso contrario, se concluye que ha sido muy bajo y se itera nuevamente, comenzando en la etapa 1

Nota: ya que la condición terminal es el nivel de estado estacionario, se recomienda usar un valor de $T$ relativamente grande. Si las gráficas obtenidas contienen una sección plana al finalizar (sección constante en el estado estacionario), se estará seguro que se ha encontrado la solución

```python
import numpy as np
import matplotlib.pyplot as plt
from copy import copy
from mpmath import mp

mp.dps = 100
```

```python
# Ejemplo: Parámetros a utilizar

α = mp.mpf(0.3)
β = mp.mpf(0.8)
θ = mp.mpf(0.6)
δ = mp.mpf(0.2)
gA = mp.mpf(1 + 0.1)
gL = mp.mpf(1 + 0.2)
τC = mp.mpf(0.19)
τY = mp.mpf(0.27)

kstar = ((α*β*(1-τY))/((gA**θ) + β*δ - β))**(1/(1-α)) # k de EE
cstar = (1/(1+τC)) * ((kstar**α)*(1-τY) + (1-δ-gA*gL)*kstar)  # c de EE

k_max = ((1-τY)/(gA*gL+δ-1))**(1/(1-α)) # k máximo (c = 0)
```

```python
k0 = ((α*β)/((gA**θ) + β*δ - β))**(1/(1-α)) # k de EE sin impuestos

# Conjetura inicial para el consumo
cL = mp.mpf(0.01) # Un valor pequeño mayor que cero
cH = mp.mpf(cstar*4) # Un valor alto, mayor que el de EE

# Parámetros del método de solución
T = 20
tol = 1e-4
maxiter = 500
flag = 0
k = np.zeros(T, dtype=object)
c = np.zeros(T, dtype=object)

for i in range(maxiter):

    c0 = (cL+cH)/2
    c[0] = c0
    k[0] = k0

    for t in range(T-1):
        k[t+1] = ( (1-τY)*k[t]**α + (1-δ)*k[t] - (1+τC)*c[t] ) / ( gA*gL )
        
        if k[t+1] < 0:
            flag = 1 # Consumo inicial muy alto
            break
        elif k[t+1] > k_max:
            flag = 2 # Consumo inicial muy bajo
            break
        else:
            c[t+1] = ((c[t]**θ)*β*(gA**(-θ))*(α*(1-τY)*(k[t+1]**(α-1)) + 1 - δ))**(1/θ) # Evolución según ecuación de Euler
        
        if t == T-2:
            if abs(k[T-1]-kstar)/kstar < tol and abs(c[T-1]-cstar)/cstar < tol:
                flag = 10
            elif k[T-1]-kstar < -tol:
                flag = 1 # Consumo inicial muy alto
            elif k[T-1]-kstar > tol:
                flag = 2 # Consumo inicial muy bajo

    if flag == 1:
        cH = copy(c0) # Eliminamos la parte alta del intervalo
    elif flag == 2:
        cL = copy(c0) # Eliminamos la parte bajo del intervalo
    elif flag == 10:
        print("Solución encontrada en {i} iteraciones".format(i=i))
        print("El consumo inicial es: {c0}".format(c0=float(c0)))
        break
```

```python
plt.figure(figsize=(8,6))
plt.plot(k, c = "steelblue", label = "Evolución Capital", linewidth=3)
plt.plot(c, c = "red", label = "Evolución Consumo", linewidth=3)
plt.xlabel("Periodos") ; plt.ylabel("Valores") ; plt.title("Evolución de variables")
plt.legend() ; plt.grid() ; plt.xticks(range(0,20))
plt.show()
```
